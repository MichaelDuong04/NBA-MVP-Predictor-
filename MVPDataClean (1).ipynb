```python
def single_team(df):
    if df.shape[0] == 1:
        return df
    else:
        # Use .apply() with a lambda that checks if the value ends with "TM"
        row = df[df["Team"].apply(lambda x: isinstance(x, str) and x.endswith("TM"))]
        # Replace the "Team" value with the final team's name from the group
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
nicknames = {}

with open("nicknames.csv", encoding = "utf-8") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
    
```

```python
combined["Team"] = combined["Team"].map(nicknames)
```
